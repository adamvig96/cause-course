```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
np.random.seed(42069)
```

```python
n = 500

df = pd.DataFrame(
    {
        "A": np.random.normal(size=n),
        "C": np.random.normal(size=n),
    }
).assign(B=lambda df: df["A"] * 0.4 + df["C"] * 0.4 + np.random.normal(size=n))
```

```python
model = sm.OLS(df.loc[:, "C"], df.drop("C", axis=1).assign(const=1)).fit()
```

```python
model.summary()
```

```python
np.random.seed(42069)

n = 8000

df = pd.DataFrame(
    {
        "IQ": np.random.normal(100, 10, size=n),
        "is_blonde": np.random.binomial(1, 0.2, size=n),
    }
).assign(
    is_on_tv=lambda df: np.random.binomial(
        1, np.clip(0.1 + df["is_blonde"] * 0.15 + (df["IQ"] - 100) * 0.01, 0, 1)
    )
)

model = sm.OLS(df.loc[:, "IQ"], df.drop("IQ", axis=1).assign(const=1)).fit()

model.summary()
```
